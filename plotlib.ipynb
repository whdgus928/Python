```python
import matplotlib.pyplot as plt
```

```python
#1부터 100까지 생성
x=range(100)
y=range(100)
plt.plot(x,y)
```

```python
import numpy as np
x_1=range(100)
y_1=[np.cos(value) for value in x_1] #cos 그래프 생성

x_2=range(100)
y_2=[np.sin(value) for value in x_2] # sin 그래프 생성

plt.plot(x_1,y_1)
plt.plot(x_2,y_2)
plt.show()
```

```python
#show를 안하면 정보가 뜸
plt.plot(x_1,y_1)
plt.plot(x_2,y_2)
```

```python
fig,ax=plt.subplots()

x_3=range(100)
y_3=[np.cos(value) for value in x_3]

ax.plot(x_3,y_3)
ax.set(title='cos graph',  #제목
      xlabel='X',  #라벨
      ylabel='Y');

plt.show()
```
